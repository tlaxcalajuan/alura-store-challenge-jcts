### _Importación de datos_



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda3.head()
```

# **Análisis de datos**

### 1. Análisis de facturación



```python
total_1 = tienda['Precio'].sum()
total_2 = tienda2['Precio'].sum()
total_3 = tienda3['Precio'].sum()
total_4 = tienda4['Precio'].sum()

#print('Total 1 = $ ' + str(total_1) + '\nTotal 2 = $ ' + str(total_2) + '\nTotal 3 = $ ' + str(total_3) + '\nTotal 4 = $ ' + str(total_4))
```

```python
ingreso_total = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingreso Total': [
        total_1, total_2, total_3, total_4
     ]
})

#print(str(total_1))
#print(str(total_2))
#print(str(total_3))
#print(str(total_4))

#ingreso_total_sum = total_1 + total_2 + total_3 + total_4
#print('Total de todas las tiendas: $ ' + str(ingreso_total_sum))

ingreso_total
```

### 2. Ventas por categoría

```python
sell_by_category_1 = tienda.groupby('Categoría del Producto').size()
sell_by_category_2 = tienda2.groupby('Categoría del Producto').size()
sell_by_category_3 = tienda3.groupby('Categoría del Producto').size()
sell_by_category_4 = tienda4.groupby('Categoría del Producto').size()

most_populars_1 = sell_by_category_1.sort_values(ascending=False).head()
most_populars_2 = sell_by_category_2.sort_values(ascending=False).head()
most_populars_3 = sell_by_category_3.sort_values(ascending=False).head()
most_populars_4 = sell_by_category_4.sort_values(ascending=False).head()

most_popular1 = sell_by_category_1.idxmax()
most_popular2 = sell_by_category_2.idxmax()
most_popular3 = sell_by_category_3.idxmax()
most_popular4 = sell_by_category_4.idxmax()
```

```python

categoria_popular = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Categoria popular': [
        most_popular1, most_popular2, most_popular3, most_popular4
     ]
})

#print('\n\nmas spopulares de 1:\n' + str(most_populars_1))
#print('\n\nmas spopulares de 2:\n' + str(most_populars_2))
#print('\n\nmas spopulares de 3:\n' + str(most_populars_3))
#print('\n\nmas spopulares de 4:\n' + str(most_populars_4))

categoria_popular
```

### 3. Calificación promedio de la tienda


```python
calificacion_promedio1 = float(tienda['Calificación'].mean())
calificacion_promedio2 = float(tienda2['Calificación'].mean())
calificacion_promedio3 = float(tienda3['Calificación'].mean())
calificacion_promedio4 = float(tienda4['Calificación'].mean())

#print('Calificacion promedio 1: ' + str(calificacion_promedio1))
#print('Calificacion promedio 2: ' + str(calificacion_promedio2))
#print('Calificacion promedio 3: ' + str(calificacion_promedio3))
#print('Calificacion promedio 4: ' + str(calificacion_promedio4))
```

```python
calificaciones_promedio = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificacion Promedio': [
        calificacion_promedio1, calificacion_promedio2, calificacion_promedio3, calificacion_promedio4
     ]
})

calificaciones_promedio
```

### 4. Productos más y menos vendidos

```python
sell_by_product_1 = tienda.groupby('Producto').size()
sell_by_product_2 = tienda2.groupby('Producto').size()
sell_by_product_3 = tienda3.groupby('Producto').size()
sell_by_product_4 = tienda4.groupby('Producto').size()

product_most_popular1 = sell_by_product_1.idxmax()
product_most_popular2 = sell_by_product_2.idxmax()
product_most_popular3 = sell_by_product_3.idxmax()
product_most_popular4 = sell_by_product_4.idxmax()

product_less_popular1 = sell_by_product_1.idxmin()
product_less_popular2 = sell_by_product_2.idxmin()
product_less_popular3 = sell_by_product_3.idxmin()
product_less_popular4 = sell_by_product_4.idxmin()

df_products = pd.concat(
    [
        sell_by_product_1.rename('Tienda 1'),
        sell_by_product_2.rename('Tienda 2'),
        sell_by_product_3.rename('Tienda 3'),
        sell_by_product_4.rename('Tienda 4')
    ],
    axis=1
).fillna(0)
```

```python
ventas_por_productos = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Mas vendido': [
        product_most_popular1, product_most_popular2, product_most_popular3, product_most_popular4
     ],
    'Menos vendido': [
        product_less_popular1, product_less_popular2, product_less_popular3, product_less_popular4
     ]
})

ventas_por_productos
```

### 5. Envío promedio por tienda

```python
costo_envio_promedio1 = float(tienda['Costo de envío'].mean())
costo_envio_promedio2 = float(tienda2['Costo de envío'].mean())
costo_envio_promedio3 = float(tienda3['Costo de envío'].mean())
costo_envio_promedio4 = float(tienda4['Costo de envío'].mean())
```

```python
coste_envio_promedio = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Costo envio promedio': [
        costo_envio_promedio1, costo_envio_promedio2, costo_envio_promedio3, costo_envio_promedio4
     ]
})

coste_envio_promedio
```

### _Uniendo Análisis_

```python
df_analisis_final = pd.concat(
    [
        ingreso_total.set_index("Tienda"),
        categoria_popular.set_index("Tienda"),
        calificaciones_promedio.set_index("Tienda"),
        ventas_por_productos.set_index("Tienda"),
        coste_envio_promedio.set_index("Tienda")
    ],
    axis=1
).reset_index()

df_analisis_final
```

# **Informe Final — Tienda a vender**

## Introducción

El objetivo fue identificar cuál de las cuatro tiendas de Alura Store es la **menos eficiente** tomando en cuenta ingresos totales, productos vendidos, calificación promedio, categoría popular y costo de envío.

## Análisis

* **Ingresos Totales:** La **Tienda 4** registra el **ingreso más bajo** del grupo (1.038B), quedando por debajo de las otras tres tiendas.
* **Calificación Promedio:** Su calificación (3.99) es **intermedia**, no destaca hacia arriba ni hacia abajo, por lo que **no compensa** su bajo desempeño en ingresos.
* **Costo de Envío:** Tiene el **costo de envío más bajo**, pero esta ventaja operativa no logra equilibrar su resultado general.
* **Categoría Popular:** Todas las tiendas comparten la categoría “Muebles”, por lo que este factor no diferencia a ninguna ni beneficia especialmente a la Tienda 4.

## Conclusión

La tienda que debe vender el Sr. Juan es **la Tienda 4**, porque presenta **el menor ingreso total**, un desempeño **promedio en satisfacción**, y **sin factores sobresalientes** que compensen su bajo resultado financiero. Su posición global la convierte en la opción menos eficiente dentro de las cuatro.
